```python
fifa =pd.read_csv('players_20.csv')
```

```python
fifa['nationality'].value_counts()
```

```python
#Top 10 nationalities that players belong to
fifa['nationality'].value_counts()[0:10]
```

```python
#Top 5 ntionalities
fifa['nationality'].value_counts()[0:5]
```

```python
#bar plot using plt function
plt.figure(figsize=(8,5))
plt.bar(list(fifa['nationality'].value_counts()[0:5].keys()), list(fifa['nationality'].value_counts()[0:5]),color="r")
plt.show()
```

```python
#creating the DataFrame player_salary
player_salary = fifa[['short_name', 'wage_eur']]
print(player_salary)
```

```python
player_salary.head()
```

```python
#Sorting the salaries of the players
player_salary = player_salary.sort_values(by=['wage_eur'], ascending = False)
```

```python
#Top 5 most earning players
player_salary.head()
```

```python
plt.figure(figsize=(8,5))
plt.bar(list(player_salary['short_name'])[0:5], list(player_salary['wage_eur'])[0:5], color="blue")
plt.show()
```
